## Importação das bibliotecas Python

A linguagem de programação utilizada em nossos experimentos será a linguagem Python. Para execução das tarefas relacionadas EDA exitem várias bibliotecas em Python que facilitam o desenvolvimento de aplicações e obtenção de informações a partir dos dados.

3.   [**Numpy**](https://numpy.org/): Biblioteca Python com licença tipo BDS para manipulação de arrays e matrizes. É uma ferramenta extremamente útil para trabalhar a base de dados.
4. [**Seaborn**](https://github.com/mwaskom/seaborn): É uma biblioteca Python também com licença tipo BSD utilizada para visualização de dados em alto nível. Trabalha com a biblioteca **Matplotlib** e possui integração com as estruturas da biblioteca **Pandas**.
5. [**Scikit-learn**](https://scikit-learn.org/stable/): Trata-se de uma biblioteca de aprendizado de máquina escrita em Pyton, possui licença licença tipo BSD e inclui vários algoritmos de classificação, regressão e agrupamento. É uma das bibliotecas mais utilizadas no desenvovimento de algoritmos de aprendizado de máquina.

```
# Importanto as bibliotecas Python necessarias ao experimento
# Manipulacao matricial e visualizacao grafica
#--------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Carregando a base de dados em um objeto DataFrame pertencente a biblioteca Pandas
# O arquivo .csv esta disponivel para download no Google Classroom.
localDataEndereco = "/content/mtcars.csv"
df = pd.read_csv(localDataEndereco)
```

O primeiro passo de qualquer EDA é saber a dimensão da base:

```
# Exibindo  dimensao da base de dados
print("Dimensão dos dados:")
print("Linhas (Instâncias):{}".format(df.shape[0]))
print("Colunas (Atributos):{}".format(df.shape[1]))
```

Sempre que possível é importante observar os valores da base de dados pois permite a inspeção e obtenção de conhecimento sobre os dados.

Como se trata de um base de apenas `32` instâncias (linhas) e `11` atributos (colunas) é possível analisar os números em pouco tempo.

```
# Apresenta, de forma completa, as cinco primeiras instâncias da base de dados.
df.head()
```

A base MTCars é formada por `32` modelos de carros de várias montadoras.

Uma análise interessante consiste em separar os carros por fabricante. Para isso vamos adicionar uma coluna extra contendo a marca de cada veículo.

```
#Criando uma coluna adicional para classificar os modelos por marca:
df['maker'] = df.name.apply(lambda x: x.split()[0])
df['maker']
```

```
# Apresentando novamente os dados com a nova coluna:
df.head()
```

Com a inclusão da coluna das marcas é possível realizar o cálculo de algumas métricas com separação por marca.

Um valor interessante é o calculo do desempenho médio por marca em termos do `mpg` médio:

```
#Criando a variavel av_mpg que apresenta o valor médio de mpg por montadora:
av_mpg  =  df.groupby ( 'maker' ).mpg.mean()
av_mpg
```

Apresentando os dados de `mpg` médio por marca em uma visão gráfica:

```
#Comando para realizar uma exibição gráfica da variavel av_mpg:

plt.figure(figsize=(3, 8))
ax = plt.gca()
av_mpg2 = av_mpg.sort_values()
makes = av_mpg2.index
speeds = av_mpg2.values.round(4)
nums = np.arange(len(av_mpg2))
plt.barh(nums, speeds)
for p, c, ch in zip(nums, makes, speeds):
    plt.annotate(str(ch), xy=(ch + 1, p), va='center')
ticks = plt.yticks(nums, makes)
xt = plt.xticks()[0]
plt.xticks(xt, [' '] * len(xt))
plt.grid(axis = 'x', color ='white', linestyle='-')
ax.tick_params(axis='both', which='both',length=0)
sns.despine(left=True, bottom=True)
```

Em uma outra situação é possível apresentar o número de carro que corresponde a cada média de `mpg`.

```
#Comando para apresentar o numero de veículos em cada av_mpg:

with sns.axes_style("whitegrid"):
    plt.hist(df.mpg.values, bins=20)
    plt.xlim(5, 40)
    plt.ylim([0, 8])
    plt.axvline(df.mpg.mean(), 0, 0.75, color='r', label='Mean')
    plt.xlabel("mpg")
    plt.ylabel("Número de carros")
    plt.title("Histograma: Milhas por galão")
    plt.legend()
```

Outra análise interessante é realizar o comparativo do desempenho em termos de `mpg` levando em consideração o tipo de transmissão do carro (`am`).

```
#Apresentando um grafico boxplot para analisar o desempenho dos carros com
#transmissão manual (1) e automática (0):

sns.boxplot(x='am', y='mpg', data=df);
```

Em uma outra análise mostramos a relação entre a variável `mpg` com o número de cilindros (`cyl`) do carro e o tipo de transmissão.

```
#Relacionando as variáveis mpg, cyl e am em um gráfico do tipo violino:

sns.violinplot(x='cyl', y='mpg', hue='am', order=[4, 6, 8], data=df, split=True);
```

Utilizando a biblioteca pandas podemos apresentar a relação entre `cyl` e `am` em formato tabular.

```
#Relacionando as variáveis cyl e am em uma tabela:
pd.crosstab(df.am, df.cyl)
```

Relacionando as variáveis peso (`wt`) com a variável milhas por galão (`mgp`) em um gráfico de pontos.

```
#Comando para relacionar variáveis em um gráfico de pontos:

g = sns.FacetGrid(df, col="cyl", hue="am", palette="Set1")
g.map(plt.scatter, "mpg", "wt", alpha=0.5, s=10);
```

Analisando a relação entre as variáveis através de um gráfico do tipo PairPlot.

```
#Criando um gráfico PairPlot com as vaiávies mpg

g = sns.PairGrid(df, vars=['mpg', 'hp', 'cyl', 'wt', 'qsec', 'disp'], hue="am")
g.map_diag(sns.kdeplot)
g.map_offdiag(plt.scatter, s=15)
```

O coeficiente de correlação de Pearson é um parâmetro adequado para medir a relação entre as variáveis.

```
#Calculando o coeificiente de correlação entre algumas variáveis da base de dados;

df[['mpg', 'wt', 'hp', 'qsec', 'disp']].corr()
```

Apresentando o coeficiente de correlação de forma gráfica.

```
#Apresentando a correlação através de um mapa de calor.
f, ax = plt.subplots(figsize=(15, 15))
plt.title('Correção de Pearson - HeatMap')

sns.heatmap(df.corr(),linewidths=0.7,vmax=1.0, square=True, annot=True)
```

Aprimorando a exibição de gráfico de pares com a sinalização da tendência.

```
#Grafico PairPlot com sinalização da tendência.

sns.pairplot(df, x_vars=['disp','cyl', 'hp', 'drat', 'wt'], y_vars=["mpg"],
             height=5, aspect=.8, kind="reg");
```

O gráfico BoxPlot é uma ferramenta interessante para identificar pontos que estão muito fora da média.

```
#Apresentando gráficos BoxPlot de algumas variáveis em relação a mpg.

fig, (ax1, ax2, ax3,ax4,ax5) = plt.subplots(1,5,figsize=(20,7))

fig.suptitle("Gráfico BoxPlot para algumas variáveis x mpg -  Mtcars", fontsize=35)

sns.boxplot(x="gear", y="mpg", data=df,ax=ax1)
sns.boxplot(x="carb", y="mpg", data=df,ax=ax2)
sns.boxplot(x="vs", y="mpg", data=df,ax=ax3)
sns.boxplot(x="am", y="mpg", data=df,ax=ax4)
sns.boxplot(x="cyl", y="mpg", data=df,ax=ax5)
```

Apresentando a contagem em relação a algumas variáveis.

```
#Apresentando o gráfico Countplot para algumas variáveis:

fig, (ax1, ax2, ax3,ax4,ax5) = plt.subplots(1,5,figsize=(25,7))
fig.suptitle("Countplot de variáveis da base Mtcars", fontsize=35)
sns.countplot(x="gear", data=df,ax=ax1)
sns.countplot(x="cyl", data=df,ax=ax2)
sns.countplot(x="carb", data=df,ax=ax3)
sns.countplot(x="vs", data=df,ax=ax4)
sns.countplot(x="am", data=df,ax=ax5)
```

Apresentando o gráfico PairGrid em relação a variáveis da base de dados.

```
#Apresentando o gráfico PairGrid em relação a algumas variáveis
#da base de dados.

g = sns.PairGrid(df[[ 'mpg',  'disp', 'hp', 'drat', 'wt','qsec']])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```
